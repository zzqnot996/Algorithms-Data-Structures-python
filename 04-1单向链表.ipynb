```python
class SingleNode(object):
  # 单链表的结点”
  def __init__(self,item):
    # item存放数据元素
    self.item = item
    # next是下一个节点的标识
    self.next = None

class SingleLinkList:
  def __init__(self):
    self.__head: SingleNode  = None

  # - is_empty() 链表是否为空
  def is_empty(self):
      return self.__head is None


  # - Length() 链表长度
  def length(self):
      cur = self.__head
      count = 0
      while cur is not None:
            cur = cur.next
            count += 1
      return count
      

  # - traveL() 遍历整个链表
  def travel(self):
    cur: SingleNode = self.__head
    while cur is not None:
      print(cur.item,end = "\t")
      cur = cur.next
    print("")



  # - add(item) 链表头部添加元素
  def add(self,item):
    node = SingleNode(item)
    node.next = self.__head
    self.__head = node

  # - append(item) 链表尾部添加元素
  def append(self,item):
    node = SingleNode(item)
    # 如果最后结尾没有节点 新添加的直接指向该节点
    if self.__head is None:
      self.__head = node
    # 如果有该节点
    else:
      # 遍历
      cur = self.__head
      while cur.next is not None:
        cur = cur.next # 获取最后一个节点
      cur.next = node

  # - insert(pos，item) 指定位置添加元素
  def insert(self,pos,item):
        if pos <= 0 :    
          self.add(item)
        elif pos >= self.length():
          self.append(item)
        else:
          cur = self.__head
          count = 0
          while count < (pos-1):
                count += 1
                cur = cur.next
          # 等于要插入的位置 哦一个节点ya
          node = SingleNode(item)
          # 当前的下一个是插入node的下一个
          node.next = cur.next
          # 节点是当前的下一个
          cur.next= node



  # - remove(item) 删除节点
  def remove(self,item):
        cur: SingleNode = self.__head
        pre = None
        while cur is not None:
            if cur.item == item:
                # 在头部 直接覆盖掉
                if cur == self.__head:
                    self.__head = cur.next
                else:
                    pre.next = cur.next
                return True
            pre = cur # 记录一下上一个节点，开启下一个轮回
            cur = cur.next
        



  # - search(item) 查找节点是否存在
  def search(self,item):
        cur: SingleNode = self.__head
        while cur is not None:
            if cur.item == item:
                return True
            cur = cur.next
        return False

if __name__ == "__main__":

    single_link_list = SingleLinkList()
    single_link_list.add("元素1")
    single_link_list.add("元素2")
    single_link_list.add("元素3")
    single_link_list.append("元素4")
    single_link_list.append("元素5")
    single_link_list.travel()
    print(single_link_list.search("元素6"))
    single_link_list.remove("元素1")
    single_link_list.travel()
    single_link_list.insert(12,"元素6666")
    single_link_list.travel()
    print(single_link_list.length())
    print(single_link_list.is_empty())
```
