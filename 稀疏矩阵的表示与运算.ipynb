## 三.稀疏矩阵的代码定义

### 1.稀疏矩阵的元素

对于矩阵的一个元素，我们需要存在值value，行下标row，列下表col。那么就定义一个结构体，并开辟三个整数(为了方便，其实值也可以是一个结构体)。

### 2.稀疏矩阵的定义

一个矩阵，我们肯定要有矩阵的行大小和列大小，且稀疏矩阵因该还有变量存非零元素的个数。
那么我们还是动态生成一段空间储存元素，然后三个变量:mrow mcol nzero
其中我们规定:0 <= row < mrow 0 <= col < mcol。

```python
import numpy as np

# 初始化矩阵
def MatInit(mrow, mcol, nzero, data):
    # 创建一个零矩阵
    mat = np.zeros((mrow, mcol), dtype=int)

    # 将非零元素填充到矩阵中
    for i in range(nzero):
        row, col, value = data[i]
        mat[row][col] = value

    return mat

# 显示矩阵
def MatShow(mat):
    for row in mat:
        print(" ".join(map(str, row)))

# 矩阵加法
def MatOP(MatA, MatB):
    return MatA + MatB

# 矩阵转置
def MatTrans(Mat):
    return Mat.T

if __name__ == "__main__":
    # 初始化矩阵
    mrow = 3
    mcol = 3
    nzero = 4
    data = [(0, 0, 1), (1, 1, 2), (2, 0, 3), (2, 2, 4)]
    mat1 = MatInit(mrow, mcol, nzero, data)
    print("Matrix 1:")
    MatShow(mat1)

    # 进行矩阵转置
    mat4 = MatTrans(mat1)
    print("\nTransposed Matrix:")
    MatShow(mat4)

```
