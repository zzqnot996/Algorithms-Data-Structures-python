```python
class SingleNode(object):
      # 单链表的结点”
  def __init__(self,item):
    # item存放数据元素
    self.item = item
    # next是下一个节点的标识
    self.next = None

class CycleSingleLinkList:
  def __init__(self):
    self.__head: SingleNode  = None

  # - is_empty() 链表是否为空
  def is_empty(self):
      return self.__head is None


  # - Length() 链表长度
  def length(self):
      if  self.is_empty():
            return 0
      
      cur = self.__head
      count = 1
      while cur.next != self.__head :
            
            count += 1
            cur = cur.next
      return count
      

  # - traveL() 遍历整个链表
  def travel(self):
    if self.is_empty():
       print("")
       return

    cur = self.__head

    while  cur.next != self.__head:
       print(cur.item, end = "")
       cur = cur.next
    # 从循环退出，cur指向尾节点
    print(cur.item)




  # - add(item) 链表头部添加元素
  def add(self,item):
    '''
    :param item: 要保存的具体数据
    '''
    node = SingleNode(item)

    if self.is_empty():
       self.__head = node
       node.next = node
    # 寻找尾节点
    cur = self.__head
    while cur.next != self.__head:
       cur = cur.next
    # 从循环退出 cur指向尾系节点  换self.__head
    node.next = self.__head
    self.__head = node
    cur.next = self.__head

  # - append(item) 链表尾部添加元素
  def append(self,item):
    node = SingleNode(item)
    if self.is_empty():
       self.__head = node
       node.next = node
    else:
    
      # 遍历
      cur = self.__head
      while cur.next != self.__head:
        cur = cur.next # 获取最后一个节点
      cur.next = node
      node.next = self.__head

  # - insert(pos，item) 指定位置添加元素
  def insert(self,pos,item):
        if pos <= 0 :    
          self.add(item)
        elif pos >= self.length():
          self.append(item)
        else:
          cur = self.__head
          count = 0
          while count < (pos-1):
                count += 1
                cur = cur.next
          # 等于要插入的位置 哦一个节点ya
          node = SingleNode(item)
          # 当前的下一个是插入node的下一个
          node.next = cur.next
          # 节点是当前的下一个
          cur.next= node



  # - remove(item) 删除节点
  def remove(self,item):
        
        if self.is_empty():
           return
        cur: SingleNode = self.__head
        pre = None
        while cur.next != self.__head :
            
            if cur.item == item:
                # 在头部 直接覆盖掉
                if cur == self.__head:
                    # 先找到尾节点
                    rear = self.__head
                    while rear.next != self.__head:
                       rear = rear.next
                    # 退出循环后 rear指向尾节点
                    self.__head = cur.next
                    rear.next = self.__head
                else:
                    pre.next = cur.next
                return True
            pre = cur # 记录一下上一个节点，开启下一个轮回
            cur = cur.next
        # 退出循环后 cur指向尾节点
        if cur.item == item:
           # 链表只有一个节点
           if cur == self.__head:
                self.__head = None
           else:
                pre.next = self.__head
              
        



  # - search(item) 查找节点是否存在
  def search(self,item):
        if self.is_empty():
               return False
        cur: SingleNode = self.__head
        while cur.next != self.__head :
        
            if cur.item == item:
                return True
            cur = cur.next
        # 退出循环后 cur指向尾节点
        if cur.item == item:
            return True
        return False
  


if __name__ == "__main__":

    single_link_list = CycleSingleLinkList()
    single_link_list.add("元素1")
    single_link_list.add("元素2")
    single_link_list.add("元素3")
    single_link_list.append("元素4")
    single_link_list.append("元素5")
    single_link_list.travel()
    print(single_link_list.search("元素6"))
    single_link_list.remove("元素1")
    single_link_list.travel()
    single_link_list.insert(12,"元素6666")
    single_link_list.travel()
    single_link_list.remove("元素3")
    single_link_list.travel()
    print(single_link_list.length())
    print(single_link_list.length())
    # 
```
