```python
class BiTreeNode:
  def __init__(self,data):

   self. data = data
   self.lchild = None
  #左孩子
   self.rchild =None#右孩子

a = BiTreeNode("A")
b = BiTreeNode("B")
c = BiTreeNode("c")
d = BiTreeNode("D")
e = BiTreeNode("E")
f = BiTreeNode("F")
g = BiTreeNode("G")

e.lchild = a
e.rchild = g
a.rchild = c
c.lchild = b
c.rchild = d
g.rchild = f

root = e

print(root.lchild.rchild.data)
```
